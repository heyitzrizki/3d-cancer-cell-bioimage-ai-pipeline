# 0.1 Import Libraries

```python
from pathlib import Path

import torch

from bioimage_pipeline.deep_learning import CTCAnnotatedPlaneDataset, UNet2D
```

# 0.2 Set Paths

```python
PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
DATASET_DIR = PROJECT_ROOT / "data" / "raw" / "Fluo-C3DL-MDA231"
```

# 1.1 Build Annotated Plane Dataset

```python
dataset = CTCAnnotatedPlaneDataset(
    DATASET_DIR,
    sequence_ids=["01", "02"],
    resize=(128, 128),
)
len(dataset)
```

# 1.2 Inspect Samples

```python
image, mask, metadata = dataset[0]
image.shape, mask.shape, image.min().item(), image.max().item(), metadata
```

# 2.1 Define 2D U-Net

```python
model = UNet2D(base_channels=4)
model
```

# 2.2 Quick Forward Pass

```python
model.eval()
with torch.no_grad():
    logits = model(image.unsqueeze(0))
logits.shape
```
