# 0.1 Import Libraries

```python
from pathlib import Path

import pandas as pd

from bioimage_pipeline.tracking import (
    add_temporal_features,
    build_tracks,
    plot_speed_distribution,
    plot_track_duration_distribution,
    plot_tracks_projection,
    summarize_tracks,
)
```

# 0.2 Set Paths

```python
PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

FEATURES_FILE = PROJECT_ROOT / "reports" / "milestone_3" / "features" / "mda231_cell_features.csv"
OUTPUT_DIR = PROJECT_ROOT / "reports" / "milestone_4"
MAX_DISTANCE = 50.0
```

# 1.1 Load Cell Feature Table

```python
features = pd.read_csv(FEATURES_FILE)
features.head()
```

# 1.2 Inspect Frame-Level Objects

```python
features.groupby("frame").size().rename("object_count")
```

# 2.1 Link Objects Across Frames

```python
linked_objects = build_tracks(features, max_distance=MAX_DISTANCE)
```

# 2.2 Assign Track IDs

```python
linked_objects[["frame", "label", "track_id", "centroid_z", "centroid_y", "centroid_x"]].head()
```

# 3.1 Add Temporal Features

```python
tracks = add_temporal_features(linked_objects)
tracks.head()
```

# 3.2 Summarize Tracks

```python
track_summary = summarize_tracks(tracks)
track_summary.head()
```

# 4.1 Visualize Cell Tracks

```python
FIGURES_DIR = OUTPUT_DIR / "figures"
plot_tracks_projection(tracks, FIGURES_DIR / "cell_tracks_projection.png")
plot_speed_distribution(tracks, FIGURES_DIR / "speed_distribution.png")
plot_track_duration_distribution(
    track_summary,
    FIGURES_DIR / "track_duration_distribution.png",
)
```

# 5.1 Export Tracking Outputs

```python
TRACKS_DIR = OUTPUT_DIR / "tracks"
TRACKS_DIR.mkdir(parents=True, exist_ok=True)
tracks.to_csv(TRACKS_DIR / "mda231_cell_tracks.csv", index=False)
track_summary.to_csv(TRACKS_DIR / "mda231_track_summary.csv", index=False)
```
